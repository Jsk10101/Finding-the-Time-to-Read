```python
import pandas as pd

import os, psutil
def printmem():
    process = psutil.Process(os.getpid())
    print(round(process.memory_info().rss/(10**9),3),'Gbytes')  # in bytes 

print("Start of process, memory used:")
printmem()

df_genres = pd.read_json('/fs/ess/PAS2038/PHYSICS_5680_OSU/project_data/goodreads/goodreads/complete/goodreads_book_genres_initial.json.gz',lines=True)
print("Memory after dataframe read in:")
printmem()
print("Size of dataframe:",len(df_genres))
print(df_genres.head(5))
```

```python
print(df_genres['genres'][10])
```

```python
df_genre_romance = df_genres[df_genres['genres'].map(lambda x: 'romance' in x)]
print(len(df_genre_romance))
df_genre_history_historicalfiction_biography = df_genres[df_genres['genres'].map(lambda x: 'history, historical fiction, biography' in x)]
print(len(df_genre_history_historicalfiction_biography))
df_genre_fantasy_paranormal = df_genres[df_genres['genres'].map(lambda x: 'fantasy, paranormal' in x)]
print(len(df_genre_fantasy_paranormal))
```

```python
df_genre_youngadult = df_genres[df_genres['genres'].map(lambda x: 'young-adult' in x)]
print(len(df_genre_youngadult))
df_genre_children = df_genres[df_genres['genres'].map(lambda x: 'children' in x)]
print(len(df_genre_children))
df_genre_nonfiction = df_genres[df_genres['genres'].map(lambda x: 'non-fiction' in x)]
print(len(df_genre_nonfiction))
df_genre_comics_graphic = df_genres[df_genres['genres'].map(lambda x: 'comics, graphic' in x)]
print(len(df_genre_comics_graphic))
df_genre_poetry = df_genres[df_genres['genres'].map(lambda x: 'poetry' in x)]
print(len(df_genre_poetry))
df_genre_mystery_thriller_crime = df_genres[df_genres['genres'].map(lambda x: 'mystery, thriller, crime' in x)]
print(len(df_genre_mystery_thriller_crime))
```
