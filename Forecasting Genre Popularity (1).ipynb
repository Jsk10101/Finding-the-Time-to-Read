```python
#Plot 10 years of data
import matplotlib.pyplot as plt
import numpy as np
R_df = R_df.drop(R_df[R_df['publication_day'] > 28].index)
R_df = R_df.drop(R_df[R_df['publication_year'] > 2016].index)
R_df = R_df.drop(R_df[R_df['publication_year'] < 2007].index)
len(R_df)
R_df['publication_date'] = pd.to_datetime(dict(day=R_df['publication_day'], month=R_df['publication_month'], year=R_df['publication_year']), format = '%Y-%m-%d %H:%M')
print(R_df['publication_date'])
R_df['year_month'] = R_df['publication_date'].dt.to_period('M')
print(R_df['year_month'])
R_df['genre'] = 'Romance'
genre_popularity2 = R_df.groupby(['year_month', 'genre']).size().reset_index(name='review_count')
print(genre_popularity2)
genre_popularity_pivot2 = genre_popularity2.pivot(index='year_month', columns = 'genre', values='review_count')
genre_popularity_pivot2.plot(figsize=(10, 6))
plt.title('Popularity of Genres Over Time')
plt.ylabel('Number of Reviews')
plt.show()

# Forecast using 10 years of data
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
# Example: Select a specific genre's data (e.g., 'Fantasy')
genre_popularity2 = genre_popularity2.dropna(subset=['year_month'])
genre_data = genre_popularity2['review_count']  # Make sure this is the review count for the genre you're forecasting
# Fit ARIMA model (example: ARIMA(5,1,0) - adjust these parameters as needed)
model = ARIMA(genre_data, order=(5,1,0))
model_fit = model.fit()
# Forecast next 12 periods (months, if using monthly data)
forecast = model_fit.forecast(steps=12)

# Plotting the forecast
plt.plot(genre_data, label='Historical Data')
plt.plot(range(len(genre_data), len(genre_data) + len(forecast)), forecast, label='Forecast')
plt.title('Projection Romance Popularity')
plt.ylabel('Number of Reviews')
plt.xlabel('Months (2006-2016)')
plt.legend()
plt.show()
```
