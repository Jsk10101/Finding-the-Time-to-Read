```python
import pandas as pd

import os, psutil
def printmem():
    process = psutil.Process(os.getpid())
    print(round(process.memory_info().rss/(10**9),3),'Gbytes')  # in bytes 

print("Start of process, memory used:")
printmem()

df_works = pd.read_json('/fs/ess/PAS2038/PHYSICS_5680_OSU/project_data/goodreads/goodreads/complete/goodreads_book_works.json.gz',lines=True)
print("Memory after dataframe read in:")
printmem()
print("Size of dataframe:",len(df_works))
print(df_works.head(5))
```

### Romance Genres Only

```python
df_genre_romance = df_genres[df_genres['genres'].map(lambda x: 'romance' in x)]
print(len(df_genre_romance))
```

```python
chunks = pd.read_json('/fs/ess/PAS2038/PHYSICS_5680_OSU/project_data/goodreads/goodreads/complete/goodreads_books.json.gz', lines=True, chunksize = 100000)
```

```python
# For Romance genre
df_all = pd.DataFrame()
for chunk in chunks:
    merged_df = pd.merge(df_genre_romance, chunk, on='book_id')
    print(merged_df.head())
    df_all = pd.concat([df_all, merged_df])
df_all.to_csv('Romance_books.csv')
```
